```python
import osmnx as ox
import networkx as nx
import matplotlib.pyplot as plt
```

```python
place = "Tver, Russia"
graph = ox.graph_from_place(place, network_type="drive")
```

```python
graph = graph.to_undirected()
```

```python
assortativity = nx.degree_assortativity_coefficient(graph)
print(f"Assortativity for Tver: {assortativity:.4f}")
```

```python
fig, ax = ox.plot_graph(graph, node_size=10, edge_linewidth=0.5, show=False, close=False)
plt.title(f"Tver road graph\nAssortativity: {assortativity:.4f}")
plt.show()
```

```python
avg_degree = sum(dict(graph.degree()).values()) / graph.number_of_nodes()
print(f"avg_degree: {avg_degree:.2f}")
```

```python
degree_distribution = nx.degree_histogram(graph)
print("degree_distribution:", degree_distribution)
```
