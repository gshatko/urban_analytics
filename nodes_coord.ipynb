```python
import osmnx as ox
import networkx as nx
import matplotlib.pyplot as plt

# Загружаем дорожный граф для города Тверь
place_name = "Тверь, Россия"
graph = ox.graph_from_place(place_name, network_type='drive')

# Визуализируем граф
fig, ax = ox.plot_graph(graph, node_size=5, node_color='r', edge_linewidth=1)

# Получаем узлы (вершины) графа
nodes = graph.nodes()

# Выводим количество вершин
print(f"Количество вершин в графе: {len(nodes)}")

# Пример распределения вершин по координатам
for node_id, data in nodes(data=True):
    x, y = data['x'], data['y']
    print(f"Вершина {node_id} имеет координаты: ({x}, {y})")

# Если нужно сохранить граф в файл
ox.save_graphml(graph, filepath='tver_graph.graphml')

# Если нужно загрузить граф из файла
# graph = ox.load_graphml('tver_graph.graphml')

# Дополнительно: можно построить граф с помощью networkx
pos = {node: (data['x'], data['y']) for node, data in nodes(data=True)}
nx.draw(graph, pos, node_size=5, node_color='r', edge_color='b', width=1)
plt.show()
```
