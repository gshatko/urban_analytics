```python
import osmnx as ox
import collections
import matplotlib.pyplot as plt

place_name = "Tver, Russia"
graph = ox.graph_from_place(place_name, network_type='drive')

degrees = [degree for node, degree in graph.degree()]

degree_count = collections.Counter(degrees)

sorted_degree_count = sorted(degree_count.items())

print("Распределение степеней вершин:")
for degree, count in sorted_degree_count:
    print(f"Degree {degree}: {count} вершин")

# Визуализируем распределение степеней
plt.figure(figsize=(10, 6))
plt.bar(degree_count.keys(), degree_count.values(), color='skyblue')
plt.title('Распределение степеней вершин в графе Твери')
plt.xlabel('Степень вершины (Degree)')
plt.ylabel('Количество вершин')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
